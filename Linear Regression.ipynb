```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
#Nama dataframe kita adalah df yang berisi data dari data.csv
#Hanya menggunakan kolom horsepower dan price
df = pd.read_csv('clustering.csv', usecols=['Unit Cost','Total'])
```

```python
x = df['Unit Cost'].values.reshape(-1,1)
y = df['Total'].values.reshape(-1,1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
lin_reg = LinearRegression()
```

```python
lin_reg.fit(x_train, y_train)
```

```python
print(lin_reg.coef_)
print(lin_reg.intercept_)
```

```python
lin_reg.score(x_test, y_test)
```

```python
y_prediksi = lin_reg.predict(x_test)
plt.scatter(x_test, y_test)
plt.plot(x_test, y_prediksi, c='r')
plt.xlabel('Unit Cost')
plt.ylabel('Total')
plt.title('Plot Unit Cost vs Total')
```
